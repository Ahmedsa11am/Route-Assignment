```python

class Bike:
    def __init__(self, name, price):
        self._name = name
        self._price = price

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        self._name = name

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, price):
        self._price = price


class ElectricBike(Bike):
    def __init__(self, name, price, maxspeed):
        super().__init__(name, price)
        self._maxspeed = maxspeed

    @property
    def maxspeed(self):
        return self._maxspeed

    @maxspeed.setter
    def maxspeed(self, maxspeed):
        self._maxspeed = maxspeed


```

```python


class NormalBike(Bike):
    def __init__(self, name, price, maxspeed):
        super().__init__(name, price)
        self._maxspeed = maxspeed

    @property
    def maxspeed(self):
        return self._maxspeed

    @maxspeed.setter
    def maxspeed(self, maxspeed):
        self._maxspeed = maxspeed
```

```python
class Customer:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        self._name = name

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, age):
        self._age = age



class MainBikeRental:
    def __init__(self):
        self._bikes = []
        self._rentals = []

    @property
    def bikes(self):
        return self._bikes

    @bikes.setter
    def bikes(self, bikes):
        self._bikes = bikes

    @property
    def rentals(self):
        return self._rentals

    @rentals.setter
    def rentals(self, rentals):
        self._rentals = rentals

    def requestBike(self, customer, number_of_bikes):
        if customer.age <= 6:
            print("Customer must be older than 6 years.")
            return

        available_bikes = len(self._bikes)
        if number_of_bikes <= available_bikes:
            self._bikes = self._bikes[number_of_bikes:]
            self._rentals.append((customer, number_of_bikes))
            print(f"{number_of_bikes} bike(s) rented to {customer.name}.")
        else:
            print("Wait 10 mins. Not enough bikes available.")

    def returnBike(self, customer, rental_time, num_of_bikes):
        self._bikes += [None] * num_of_bikes  
        cost = rental_time * 40 * num_of_bikes
        print(f"Total bill for {customer.name} is ${cost}.")
        self._rentals = [rental for rental in self._rentals if rental[0] != customer]

    def addBike(self, bike):
        self._bikes.append(bike)

    def totalCost(self):
        return sum(bike.price for bike in self._bikes if bike)


```

```python
# Example usage:
customer1 = Customer("John Doe", 25)
bike1 = NormalBike("Mountain Bike", 100, 30)
bike2 = ElectricBike("E-Bike", 150, 50)

rental_shop = MainBikeRental()
rental_shop.addBike(bike1)
rental_shop.addBike(bike2)

rental_shop.requestBike(customer1, 1)
rental_shop.returnBike(customer1, 2, 1)
print(f"Total cost of bikes in the store: ${rental_shop.totalCost()}")
```
